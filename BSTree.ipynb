```python
class Node:
    def __init__(self,value,leftNode=None,rightNode=None):
        self.value = value
        self.leftNode = leftNode
        self.rightNode = rightNode
        
    def add(self, value):
        # if node is LE to root, add to the left side of the tree
        if (value <= self.value):
            if (self.leftNode is None):
                self.leftNode = Node(value)
            else:
                self.leftNode.add(value)
        # if node is GT to root, add to the right side of the tree
        else:
            if (self.rightNode is None):
                self.rightNode = Node(value)
            else:
                self.rightNode.add(value)
        
```

```python
class BTree:
    def __init__(self, root=None):
        self.root = root
        
    def add(self, value):
        # if root = None, make this node root
        if (self.root is None):
            self.root = Node(value)
        else:
            self.root.add(value)
            
    def traverse(self):
        if (self.root is None):
            print "empty tree"
            return
        
        currentRoot = self.root
        self.walkthrough(currentRoot)
        
    def walkthrough(self, currentRoot):
        if (currentRoot.leftNode is not None):
            self.walkthrough(currentRoot.leftNode)
        print currentRoot.value
        if (currentRoot.rightNode is not None):
            self.walkthrough(currentRoot.rightNode)
            
    def search(self, value):
        # if tree is empty return false
        if (self.root is None):
            print "empty tree"
            return False
        
        currentRoot = self.root

        while (currentRoot is not None):
            if (currentRoot.value == value):
                return True
            elif (value < currentRoot.value):
                currentRoot = currentRoot.leftNode
            else:
                currentRoot = currentRoot.rightNode
                
        return False
        
    def find(self, currentRoot, value):
        if (currentRoot.value == value):
            return True
        elif (value < currentRoot.value):
            if (currentRoot.leftNode is not None):
                self.find(currentRoot.leftNode, value)

        else:
            if (currentRoot.rightNode is not None):
                self.find(currentRoot.rightNode, value)
                        
```
